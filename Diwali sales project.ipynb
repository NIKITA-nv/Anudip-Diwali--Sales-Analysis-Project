# EXPLORATRY DATA ANALYSIS


#Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x='Gender', hue='Gender',data=df, palette=['blue', 'pink']) 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount
## Group by 'Gender' and sum the 'Sales_Amount'
gender_sales = df.groupby('Gender')['Amount'].sum()
# Create a bar plot
plt.figure(figsize=(6, 4))
plt.bar(gender_sales.index, gender_sales.values, color=['blue', 'pink'])

# Add title and labels
plt.title('Gender-wise Sales Amount')
plt.xlabel('Gender')
plt.ylabel('Total Sales Amount')
```


*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

#Age

```python
 #plotting for age agegroup and it's count.
# Ensure the Age column is numeric (if not already)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Count occurrences in each AgeGroup
age_group_counts = df['Age Group'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(age_group_counts.index, age_group_counts.values, marker='o', color='b', linestyle='-', linewidth=2)

# Adding titles and labels
plt.title('Count of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

#create a barplot
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```


*From above graph we can see that most of the buyers are of age group between 26-35 years females*

#State

```python
#Total number of order from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states
# Step 1: Group by 'State' and sum the 'SalesAmount' for each state
state_sales = df.groupby('State')['Amount'].sum().reset_index()

# Step 2: Sort the states by 'SalesAmount' in descending order and select top 10
top_states = state_sales.sort_values(by='Amount', ascending=False).head(10)

# Step 3: Create a scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(top_states['State'], top_states['Amount'], color='green', alpha=0.7, s=100)

# Step 4: Add labels, title, and customize the plot
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount from Top 10 States')
plt.grid(True)
```


*From above graph we can see that most of the order and total sales amount are from Uttarprdesh Maharashtra and Karnataka respectively*

#Marital status

```python
#plotting a bar chart for marital status and it's count
marital_counts = df['Marital_Status'].value_counts()

# Step 2: Create a bar plot
plt.figure(figsize=(6, 4))
plt.bar(marital_counts.index, marital_counts.values, color=['blue', 'pink'])

# Add title and labels
plt.title('Count of Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')

# Display the plot
plt.show()
```

```python
#Marital status and gender wise sales amount
# Ensure the Sales column is numeric (if not already)
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

# Group by Marital Status and Gender, then sum the Sales
sales_data = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().unstack()

# Plot
plt.figure(figsize=(10, 6))
for status in sales_data.index:
    plt.plot(sales_data.columns, sales_data.loc[status], marker='o', label=f'Marital_Status: {status}')

# Adding titles and labels
plt.title('Sales Amount by Marital Status and Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales Amount')
plt.legend(title='Marital Status')
plt.show()
```

 
*From above graph we can see that most of the buyers are married woman and the have high purchasing power*

#Occupation

```python
#plotting a bar plot for occupation and it's count
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Occupation wise total sales amount
#Step 1: Group by 'Occupation' and sum the 'SalesAmount' for each occupation
occupation_sales = df.groupby('Occupation')['Amount'].sum().reset_index()

# Step 2: Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(occupation_sales['Occupation'], occupation_sales['Amount'], color='red')

# Step 3: Add labels and title
plt.xlabel('Occupation')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Occupation')
plt.grid(True)
```

 
*From above graph we can see that most of the buyers are working in IT, Healthcare and Aviation Secter*

#Product Category

```python
#plotting a bar plot for product category and it's count
# Step 1: Count the occurrences of each product category
category_counts = df['Product_Category'].value_counts()

# Step 2: Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values, color='pink')

# Add title and labels
plt.title('Count of Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')

# Display the plot
plt.show()
```

```python
#Product category wise sales amount
sales_age = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

#create a barplot
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category',y= 'Amount' ,data = sales_age)
```


*From above graph we can see that most of the sold product are from Food,Clothing and Electronic Gadgets*

```python
# top 10 most sold products (same thing as above)
# Step 1: Group by 'Product' and sum 'QuantitySold' for each product
product_sales = df.groupby('Product_ID')['Orders'].sum().reset_index()

# Step 2: Sort the products by 'QuantitySold' in descending order and select top 10
top_products = product_sales.sort_values(by='Orders', ascending=False).head(10)

# Step 3: Create a scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(top_products['Product_ID'], top_products['Orders'], color='orange', alpha=0.7, s=100)

# Step 4: Add labels, title, and customize plot
plt.xlabel('Product_ID')
plt.ylabel('Orders')
plt.title('Top 10 Most Sold Products')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.show()
```
